```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.studyblue.com/notes/note/n/petrobowl-questions-part-1/deck/13481356'
req = requests.get(url)
html_data = req.text
soup = BeautifulSoup(html_data,'html.parser')
```

```python
data = []
for div in soup.findAll('div',{'class':'text text-small'}): 
    data = np.append(data,div.text)
```

```python
data = data.reshape((-1,2))
data = pd.DataFrame(data,columns=['Question','Answer'])
```
